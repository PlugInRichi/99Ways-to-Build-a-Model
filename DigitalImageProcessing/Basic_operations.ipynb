```python
import cv2 as cv
from google.colab.patches import cv2_imshow

import numpy as np
```

```python
im_lenna = cv.imread("/content/drive/MyDrive/PlugInRichi/Lenna.png")
print("Shape: ", im_lenna.shape)
cv2_imshow(im_lenna)
```

```python
im_lenna_gray = cv.cvtColor(im_lenna, cv.COLOR_BGR2GRAY)
print("Shape: ", im_lenna_gray.shape)
cv2_imshow(im_lenna_gray)
```

# Filters

```python
im_lenna_blur = cv.medianBlur(im_lenna, 5)
cv2_imshow(im_lenna_blur)
```

```python
prewitt_filter = np.array([[-1, 0, 1],[-1, 0, 1],[-1, 0, 1]])
print(prewitt_filter)
im_prewitt_lenna = cv.filter2D(im_lenna, -1, prewitt_filter)
cv2_imshow(im_prewitt_lenna)
```

```python
im_lenna_f = im_lenna_blur + im_prewitt_lenna *0.5
cv2_imshow(im_lenna_f)
```

# Morphological transformations

```python
im_bacterium = cv.imread("/content/drive/MyDrive/PlugInRichi/BACTERIAS.jpg", cv.IMREAD_GRAYSCALE)
cv2_imshow(im_bacterium)
```

```python
elipse_kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (7, 7))
print(elipse_kernel)
```

```python
img_erode = cv.erode(im_bacterium, elipse_kernel, iterations=1)
cv2_imshow(img_erode)
```

```python
img_dilate = cv.dilate(im_bacterium, elipse_kernel, iterations=1)
cv2_imshow(img_dilate)
```

```python
rec_kernel = cv.getStructuringElement(cv.MORPH_RECT, (7, 7))
print(rec_kernel)
img_dilate = cv.dilate(im_bacterium, rec_kernel, iterations=1)
cv2_imshow(img_dilate)
```
